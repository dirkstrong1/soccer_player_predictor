```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fifa_df = pd.read_csv('write_data/all_players.csv')
fifa_df.head()
```

```python
clean_df = pd.read_csv('write_data/clean_all_players.csv')
```

```python
clean_df = clean_df.set_index('player_name').drop(columns=['height_(cm)', 'weight_(kg)'])
```

```python
pca_df = pca_df.drop(columns='Unnamed: 0')
```

```python
pca_df[8175:8200]
```

```python
clean_df.describe()
```

```python

plt.scatter(pca_df['pc1'],pca_df['pc2']);
```

```python
from sklearn.cluster import KMeans # import from sklearn
model = KMeans(n_clusters=4).fit(pca_df)
#k_means = KMeans(n_clusters=6) # Must set number of clusters at initialization time!
#k_means.fit(pca_df) # Run the clustering algorithm
model.cluster_centers_
```

```python
cluster_assignments = model.predict(pca_df) # Generate cluster index values for each row in df
cluster_assignments
```

```python
from sklearn.metrics import calinski_harabasz_score
```

```python
print(calinski_harabasz_score(pca_df, cluster_assignments))
```

```python
plt.figure(figsize=(15,10))
plt.scatter(pca_df['pc1'], pca_df['pc2'], c=model.labels_)
for i in range(len(model.cluster_centers_)):
    plt.scatter(model.cluster_centers_[i][0], model.cluster_centers_[i][1]);
    
```

```python
X = pca_df
distortions = []

# Calculate SSE for different K
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state = 10)
    kmeans.fit(X)
    distortions.append(kmeans.inertia_)

# Plot values of SSE
plt.figure(figsize=(15,8))
plt.subplot(121, title='Elbow curve')
plt.xlabel('k')
plt.plot(range(2, 10), distortions)
plt.grid(True)
```

```python
from scipy.cluster.hierarchy import dendrogram, ward

# use the ward() function
linkage_array = ward(X)

# Now we plot the dendrogram for the linkage_array containing the distances
# between clusters
dendrogram(linkage_array)

fig, ax = plt.subplots(figsize=(12,12))
bounds = ax.get_xbound()
ax.plot(bounds, [16, 16], '--', c='k')
ax.plot(bounds, [9, 9], '--', c='k')
ax.text(bounds[1], 16, ' 2 clusters', va='center', fontdict={'size': 12})
ax.text(bounds[1], 9, ' 3 clusters', va='center', fontdict={'size': 12})
ax.set_xlabel("Data index")
ax.set_ylabel("Cluster distance")
```
